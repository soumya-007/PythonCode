#### Different Mode of file
# mode =r read only
# mode =w write only (will overwrites file or create new)
# mode =r+ read and write only
# mode =a   will append only
# mode =w+   writing and reading (overwrites existing file and create new files )


```python
%%writefile samplefiles/my_new_File.txt
this is frist line
this is 2nd line
this is 3rd line
```

```python
with open ('samplefiles/my_new_File.txt', mode='r' ) as f:
        print(f.read())
```

```python
with open ('samplefiles/my_new_File.txt', mode='a' ) as f:
        f.write ('\nthis is fourth line')
```

```python
with open ('samplefiles/my_new_File.txt', mode='r' ) as f:
        print(f.read())
```
